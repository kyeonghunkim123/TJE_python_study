# 시도별 전출입 인구수 데이터 시각화

```python
import matplotlib.pyplot as plt
```

```python
df.info()
```

```python
# 누락값(NaN)을 앞 데이터로 채움
df = df.fillna(method='ffill') #ffill은 forward 로 fill을 함
```

```python
# 서울에서 다른 지역으로 이동한 데이터만 추출하여 정리
mask = (df['전출지별']=='서울특별시') & (df['전입지별'] != '서울특별시')
```

```python
df_seoul = df[mask]
df_seoul.head(2)
```

```python
df_seoul = df_seoul.drop(['전출지별'], axis=1)
```

```python
df_seoul.head()
```

```python
df_seoul.rename({'전입지별':'전입지'}, axis=1, inplace=True)
df_seoul.set_index('전입지', inplace=True)
```

```python
df_seoul.head(3)
```

```python
# 서울에서 경기도로 이동한 인구 데이터 값만 선택
sr_one = df_seoul.loc['경기도']
```

```python
plt.plot(sr_one.index, sr_one.values)
```

# 남북한 전력량으로 secondary_y를 그리기

```python
df = pd.read_excel('./data1/남북한발전전력량.xlsx', engine='openpyxl', convert_float=True)
#북한 데이터만 가져오기
df = df.loc[5:9]
```

```python
df.drop("전력량 (억㎾h)", axis='columns', inplace=True)
df.head(3)
```

```python
# 인덱스를 '발전 전력별'로 변경
df.set_index('발전 전력별', inplace=True)
```

```python
df = df.T
df.tail(2)
```

```python
# 증감율 (변동률) 계산
df = df.rename(columns={'합계':'총발전량'})
```

```python
df['총발전량 - 1년'] = df['총발전량'].shift(1)
df.head(2)
```

```python
df['증감율'] = ((df['총발전량'] / df['총발전량 - 1년'])-1) * 100
df.head(2)
```

```python
# 2축 그래프 그리기
ax1 = df[['수력','화력']].plot(kind='bar', figsize=(20,10), width=0.7, stacked=True)
ax2 = ax1.twinx()
ax2.plot(df.index, df.증감율, ls='--', marker='o', markersize=20, 
        color='green', label='전년대비 증감율(%)')
#ls : line style는 라인스타일
ax1.set_ylim(0, 500)
ax2.set_ylim(-50, 50)

ax1.set_xlabel('연도', size=20)
ax1.set_ylabel('발전량(억 Kwh)')
ax2.set_ylabel('전년대비 증감율')
plt.title('북한 전력 발전량(1990~2016)', size=25)
ax1.legend(loc='upper left')
plt.show()
```

```python
#히스토그램 - 단변수 데이터의 빈도수를 나타냄
df = pd.read_csv('./auto-mpg.csv', header=None)
```

```python
df.columns = ['mpg', 'cylinders', 'displacement','horsepower', 'weight',
             'acceleration', 'model year', 'origin', 'name']
```

```python
df.head(2)
```

```python
plt.style.use('classic')
#연비(mpg) 열에대한 히스토그램을 그리기
df['mpg'].plot(kind='hist', bins=10, color='coral', figsize=(10,5))
# bins=10은 mpg10~50 정의 값을 10구간으로 나눈다.
# 그래프 꾸미기
plt.title('histogram')
plt.xlabel('mpg')
plt.show()
```

```python
# 산점도 scatter
plt.style.use('Solarize_Light2')
# plt.style.available : 스타일 확인코드
df.plot(x='weight', y='mpg', kind='scatter', c='coral', s=10, figsize=(10,5))
plt.title('scatter plot mpg VS. weight')
plt.show()
```

```python
# 버블차트
plt.style.use('bmh')
# 버블차트를 그리기 위해 cylinder 개수의 상대적 비율을 계산하여 시리즈 생성
cylinder_size = df.cylinders / df.cylinders.max() * 300
# plt.style.available : 스타일 확인코드
df.plot(x='weight', y='mpg', kind='scatter', c='coral', s=cylinder_size
        , figsize=(10,5), alpha=0.3)
plt.title('scatter plot: mpg-weight-cylinders')
plt.show()
```

```python
plt.style.use('default')
#3개의 변수로 산점도 그리기
df.plot(x='weight', y='mpg', kind='scatter', c=cylinder_size, s=50, marker='+',
        figsize=(10,5), alpha=0.3, cmap='hot') 
#cmap은 color map 'viridis', 'jet', 'gray', 'hot'
plt.title('scatter plot: mpg-weight-cylinders')
plt.show()
```

```python
# pie 차트
df['count'] = 1
df_origin = df.groupby('origin').sum()
print(df_origin)
```

```python
df_origin.index = ['USA', 'EU', 'JAPAN']
df_origin['count'].plot(kind='pie', figsize=(7, 5), autopct='%1.1f%%',
                        startangle=10, #파이조각을 나누는 시작점(각도 표시)
                        colors=['chocolate', 'bisque', 'cadetblue']) #3개의 색상리스트
plt.title('Model origin', size=20)
plt.axis('equal') #파이 차트의 비율을 같게(월에 가깝게) 조정
plt.legend(labels=df_origin.index, loc='upper right')
plt.show()
```

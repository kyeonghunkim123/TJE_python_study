# 열을 행으로 보내기(pivot)

```python
df = pd.read_csv('./data1/scores.csv')
df = df.melt(id_vars='name', var_name='subject', value_name='score')
df
```

```python
def get_grade(x):
    if x >= 90 : 
        grade = 'A'
    elif x>= 80:
        grade = 'B'
    elif x>= 70:
        grade = 'C'
    elif x>= 60:
        grade = 'D'
    else:
        grade = 'F'
    return grade     
```

```python
df['grade'] = df['score'].apply(get_grade)
df = df.sort_values('name')
df
```

```python
# 데이티프레임.pivot(index = 인덱스로 사용할 컬럼, 
# columns = 컬럼으로 사용할 컬럼, values=값으로 사용할 컬럼)
# name, subject, score
df.pivot(index='name', columns='subject', values='score')
```

```python
df.pivot(index='name', columns='subject', values='grade')
```

```python
df.pivot(index='name', columns='subject', values=['score','grade'])
```

```python
df.transpose()
```

```python
df_titanic = sns.load_dataset('titanic')
df_titanic = df_titanic[['survived','pclass', 'sex', 
                         'age', 'embarked']]
df_titanic.dropna(inplace=True)
df_titanic.isnull().sum()
```

```python
#성별, 객실등급별 승선자수  count를 사용
df_titanic.pivot_table(index='sex', columns='pclass',
                      values='survived', aggfunc='count', margins=True)
# survive = [1,0,1,1,0] 5명
```

```python
#성별, 객실등급별 생존자수 sum를 사용
df_titanic.pivot_table(index='sex', columns='pclass',
                      values='survived', aggfunc='sum',
                       margins=True)
# survive = [1,0,1,1,0] 3명
```

```python
#성별, 객실등급별 생존율 
df_titanic.pivot_table(index='sex', columns='pclass',
                      values='survived', aggfunc='mean',
                       margins=True)
# survive = [1,0,1,1,0] => 3/5 
```

```python
# aggfunc = 'mean' 은 default임 
df_titanic.pivot_table(index='sex', columns='pclass',
                      values='survived',
                       margins=True)
```

# groupby

```python
# df.groupby(그룹기준컬럼).통계적용컬럼.통계함수
# .count() : 누락값을 제외한 데이터수
# .size() : 누락값을 포함한 데이터수
# .mean(), .sum(), .std(), .min(), .max()
```

```python
#객실등급(pclass) 별 승선자수 결과 집계
df_titanic1 = df_titanic.groupby('pclass').survived.count().to_frame()
df_titanic1
```

```python
#객실등급(pclass) 별 생존자수 결과 집계
df_titanic2 = df_titanic.groupby('pclass').survived.sum().to_frame()
df_titanic2
```

```python
#객실등급(pclass) 별 생존비율 결과 집계
df_titanic3 = df_titanic.groupby('pclass').survived.mean().to_frame()
df_titanic3
```

```python
df_titanic4 = pd.concat([df_titanic1, df_titanic2, df_titanic3], 
                      axis=1)
df_titanic4.columns = ['승선자수', '생존자수', '생존비율']
df_titanic4
```

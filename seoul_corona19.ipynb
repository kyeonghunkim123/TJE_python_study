# 데이터 확인 및 전처리

```python
# 불필요한 컬럼 삭제
df = df.drop(columns=['환자번호','국적', '환자정보', '조치사항', '이동경로', 
                 '등록일', '수정일', '노출여부'])
```

```python
df.info()
```

```python
#확진일 --> datetime
df['확진일'] = pd.to_datetime(df['확진일'])
df.info()
```

```python
# 지역 25개 구 
print(len(df['지역'].unique()))
print(df['지역'].nunique())
df['지역'].unique()
```

```python
df['지역'] = df['지역'].str.strip()
```

```python
print(df['지역'].nunique())
df['지역'].unique()
```

```python
df['지역'] = df['지역'].astype('category')
```

```python
df.info()
```

```python
#결측치 확인
df.isnull().sum()
```

# 구별 확진자 동향
## 확진일- 구별로 pivot_table 집계

```python
df_gu = pd.pivot_table(df, index='확진일', columns='지역', values='연번',
                      aggfunc='count', margins=True)
# df_gu = pd.pivot_table(df,  '연번', '확진일', '지역', 'count')
df_gu
```

```python
# 서울시 일별 추가확진자 동향
s_date = df_gu['All'][:-1]
```

```python
# 서울시 일별 추가확진자 많은 순으로 
s_date.sort_values(ascending=False)
```

```python
from matplotlib import font_manager, rc
import matplotlib.pyplot as plt
import matplotlib as mpl
# minus 표시
mpl.rcParams['axes.unicode_minus'] = False
plt.rc('font', family='Malgun Gothic') 
```

```python
# 서울시 일별 추가 확진자 시각화
```

```python
x = s_date.index
y = s_date.values
plt.plot(x, y)
plt.title('서울시 일별 확진자 시각화')
plt.xlabel('확진일')
plt.ylabel('확진자수')
plt.xticks(rotation=45)
plt.show()
```

```python
# 서울시 구별 누적 확진자 비교
s_gu = df_gu.loc['All'][:-1]
s_gu = s_gu.sort_values(ascending=False)
```

```python
s_gu
```

```python
#서울시 구별 누적확진자 많은 순 시각화
x = s_gu.index
y = s_gu.values
plt.figure(figsize=(5, 4))
plt.barh(x, y)
plt.title('서울시 구별 누적확진자수', size=10)
plt.show()
```

```python
# 최근일(21-09-28) 기준 지역별 확진자
s_gu1 = df_gu.iloc[-2][:-1]
s_gu1 = s_gu1.sort_values(ascending=False)
```

```python
# 최근일(21-09-28) 기준 지역별 확진자 시각화
x = s_gu1.index
y = s_gu1.values
plt.figure(figsize=(5, 4))
plt.barh(x, y)
plt.title('최근일(21-09-28) 기준 지역별 확진자', size=10)
plt.show()
```
